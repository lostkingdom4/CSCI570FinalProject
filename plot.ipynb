```python
import matplotlib.pyplot as plt

def plot_graph(x_values, y1, y2, title='Graph', x_label='X-Axis', y_label='Y-Axis'):
    """Plot a graph given two lists of data (x and y values)."""
    # Create the plot
    plt.figure(figsize=(8, 6))
    plt.plot(x_values, y1, marker='o', linestyle='-', label='optimized')
    plt.plot(x_values, y2, marker='x', linestyle='--', label='basic')

    
    # Add title and axis labels
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    
    # Optionally, enable the grid for better readability
    plt.grid(True)
    plt.legend(loc='best')

    
    # Display the graph
    plt.show()
    
def read_performance_data(file_path):
    """Read performance data from the specified file and return structured lists."""
    problem_sizes = []
    times = []
    memory_usages = []

    index = 0
    with open(file_path, 'r') as file:
        for line in file:
            line = line.strip()

            # Skip lines that are empty or contain comments
            if not line or line.startswith("#"):
                continue

            if index == 0:
                problem_sizes.append(int(line))
                index+=1
            elif index ==1:
                times.append(float(line))
                index+=1
            elif index ==2:
                memory_usages.append(float(line))
                index=0

    return problem_sizes, times, memory_usages

file_path = './skye/mem_efficient_output.txt'  # Replace this with your file path
problem_sizes, times, memory_usages = read_performance_data(file_path)

file_path = './mem_basic.txt'  # Replace this with your file path
problem_sizesb, timesb, memory_usagesb = read_performance_data(file_path)

# Now you can access each list and perform further processing
print("Problem Sizes:", len(problem_sizes))
print("Times:", len(times))
print("Memory Usages:", len(memory_usages))

# Call the plot function
plot_graph(problem_sizes, times, timesb, title='time vs m+n', x_label='Time (s)', y_label='Size')
plot_graph(problem_sizesb, memory_usages, memory_usagesb, title='memory usage vs m+n', x_label='memory usage (KB)', y_label='Size')
```
